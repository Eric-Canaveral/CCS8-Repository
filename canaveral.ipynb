```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

d = {'miles_per_week': [37,39,46,51,88,17,18,20,21,22,23,24,25,27,28,29,30,31,32,33,
                        34,38,40,42,57,68,35,36,41,43,45,47,49,50,52,53,54,55,56,58,
                        59,60,61,63,64,65,66,69,70,72,73,75,76,77,78,80,81,82,83,84,
                        85,86,87,89,91,92,93,95,96,97,98,99,100,101,102,103,104,105,
                        106,107,109,110,111,113,114,115,116,116,118,119,120,121,123,
                        124,126,62,67,74,79,90,112],
      'completed_50m_jc': ['no','no','no','no','no','no','no','no','no','no','no','no',
                           'no','no','no','no','no','no','no','no','no','no','no','no',
                           'no','no','yes','yes','yes','yes','no','yes','yes','yes','no',
                           'yes','yes','yes','yes','yes','yes','yes','yes','no','yes','yes',
                           'yes','yes','yes','yes','yes','no','yes','yes','yes','yes','yes',
                           'yes','yes','no','yes','yes','yes','yes','yes','yes','yes','no',
                           'yes','yes','yes','yes','yes','yes','yes','yes','yes','yes','yes',
                           'yes','yes','yes','yes','yes','yes','yes','yes','yes','yes','yes','yes',
                           'yes','yes','yes','yes','yes','yes','yes','yes','yes','yes',]}

#Upload the database mga bai.
df = pd.DataFrame(d)
df
finished_race= ['no', 'yes']
enc = OrdinalEncoder(categories=[finished_race])
df['completed_50m_jc'] = enc.fit_transform(df[['completed_50m_jc']])
df
```

```python
plt.scatter(df['miles_per_week'], df['completed_50m_jc'], c=df['completed_50m_jc'], cmap='rainbow')
plt.title('Scatter Plot of Logistic Regression')
plt.show()

x = df.iloc[:, 0:1]
y = df.iloc[:, 1]

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=11)
log_reg = LogisticRegression()
log_reg.fit(x_train, y_train)
y_pred = log_reg.predict(x_test)
print(y_pred)
print(log_reg.score(x_train, y_train))

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
